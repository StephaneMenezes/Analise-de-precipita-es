```python
import matplotlib.pyplot as plt #importando a biblioteca Matplotlib
import pandas as pd             #importando a biblioteca pandas
import matplotlib.gridspec as gridspec  
import numpy as np              #importando o numpy
from google.colab import files #Upar os arquivos no colab
#vamos realizar o upload
mapa1966 = files.upload()
```

```python
acoes=pd.read_csv('mapa1966.csv')
print(acoes)
```

```python
mask=(acoes.PP09>0) #Criando um arquivo onde é excluindo os zeros da coluna pp09
acoes_sem_0=acoes[mask]
acoes_sem_0
```

```python
mask2=(acoes_sem_0.PP15>0) #Criando um arquivo onde é excluindo os zeros da coluna pp15
acoes_sem_1=acoes_sem_0[mask2]
acoes_sem_1
```

```python
mask3=(acoes_sem_1.PP21>0)   #Criando um arquivo onde é excluindo os zeros da coluna pp21
acoes_sem_2=acoes_sem_1[mask3]
acoes_sem_2
```

```python
x = (acoes_sem_0['TA09'])
y = (acoes_sem_0['PP09'])
x1=(acoes_sem_1['TA15'])
y1=(acoes_sem_1['PP15'])
x2=(acoes_sem_2['TA21'])
y2=(acoes_sem_2['PP21'])
#print(x)
#print(y)
plt.scatter(x,y, color='c', label="09h") #Serve para plotar x em relação a y
plt.scatter(x1,y1, color='r', label="15h") #label e plt legend para colocar uma legenda no gráfico
plt.scatter(x2,y2, color='b', label="21h")
plt.legend()


plt.title("Temperatura X Precipitação (09h , 15h e 21h)")
plt.xlabel("Temperatura")
plt.ylabel("Precipitação")
plt.grid(True)
```
